```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_X=scaler.fit_transform(X)
scaled_X
```

```python
#Plotting the curve of inertia of kmeans clustering to identify right value of k
from sklearn.cluster import KMeans
ssd = []

for k in range(2,10):
    
    model = KMeans(n_clusters=k,n_init=10)
    
    
    model.fit(scaled_X)
    
    #Sum of squared distances of samples to their closest cluster center.
    ssd.append(model.inertia_)
plt.plot(range(2,10),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
diff_ssd = pd.Series(ssd).diff()

# Change the index to start from 2 to 9
diff_ssd.index = range(2, 10)
diff_ssd.plot(kind='bar')
```

From above graphs we can conclude by elbow method that k=4 is correct value for kmeans clustering

```python
model=KMeans(n_clusters=4,n_init=100)
cluster_labels=model.fit_predict(scaled_X)
X_label=X.copy()
print(cluster_labels)
X_label['cluster']=cluster_labels
X_label.corr()['cluster']
```

```python
# creating an emotion cluster based on mean of features and their dependency on classes
emotion_cluster={'angry':2,'disgust':0,'happy':1,'surprise':1,'sad':0,'fear':2,'neutral':3}
```

```python
# give cluster no.s to records in main dataset
df1['cluster']=X_label['cluster']
df1['cluster'].value_counts()
```

```python
df1['Popularity'].value_counts()
```

As we can see many records have popularity close to zero. so these songs will not be appropriate to reccomend. now adding a filter that songs with popularity more than 50 only are allowed

```python
df1=df1[df1['Popularity']>50]
df1['Popularity'].value_counts()
```

```python
# saving the dataset to csv file
df1.to_csv("clustered_songs4.csv")
```

```python
#creating recommend function that takes input emotion cluster and emotion and returns recommended songs
emotion="happy"
def recommend(emotion,ec,n):
    cluster=ec[emotion]
    all_songs=df1[df1['cluster']==cluster]
    return all_songs.sample(n)
recommend(emotion,emotion_cluster,5)['Track Name']
```

```python
# creating a function named tracks that reccomend 20 songs by taking mood dictionary as input
def tracks(moods):
    if(len(moods)==0):
        return []
    for i in moods:
        moods[i]*=20
        moods[i]=round(moods[i])
    tracks=[]
    for mood in moods:
        songs=recommend(mood,emotion_cluster,moods[mood])
        cols=["Track URI","Track Name","Album Name","Album Release Date",'Popularity',"Artist Name(s)"]
        ar=np.array(songs[cols])
        out=[]
        for i in ar:
            cur={}
            for j in range(len(cols)):
                cur[cols[j]]=i[j]
            out.append(cur)
        tracks=tracks+out
    return tracks
# taking example dictionary with probabilities for different moods
moods={"happy":0.6312,"sad":0.2131,"neutral":0.1557}
tracks(moods)
```
